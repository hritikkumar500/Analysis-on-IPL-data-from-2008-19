```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# Batting Career

```python
rcb_batting=data_del[data_del['batting_team']=='Royal Challengers Bangalore']
```

```python
rcb_batting
```

```python
rcb_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rcb_batting['batsman_runs'].value_counts()[0:5].keys()),list(rcb_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS BY RCB')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rcb_batting['batsman_runs'].value_counts()),labels=list(rcb_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
rcb_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rcb_batting['dismissal_kind'].value_counts()),labels=list(rcb_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
rcb_bowling['wide_runs'].value_counts()
```

```python
rcb_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rcb_bowling['batsman_runs'].value_counts()[0:5].keys()),list(rcb_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY RCB BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball_rcb=rcb_bowling[rcb_bowling['wide_runs']!=0]
```

```python
wide_ball_rcb
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_rcb['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_rcb['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('Count Of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY RCB')
plt.xticks(rotation=45)
plt.show()
```
