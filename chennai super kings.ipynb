```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
csk_batting=data_del[data_del['batting_team']=='Chennai Super Kings']
```

```python
csk_batting
```

```python
csk_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(csk_batting['batsman_runs'].value_counts()[0:5].keys()),list(csk_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by CSK')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(csk_batting['batsman_runs'].value_counts()),labels=list(csk_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
csk_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(csk_batting['dismissal_kind'].value_counts()),labels=list(csk_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
csk_bowling
```

```python
csk_bowling['wide_runs'].value_counts()
```

```python
csk_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(csk_bowling['batsman_runs'].value_counts()[0:5].keys()),list(csk_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY CSK BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball_csk=csk_bowling[csk_bowling['wide_runs']!=0]
```

```python
wide_ball_csk
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_csk['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_csk['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('COUNT OF WIDE BALLS')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY DECCAN CHARGERS')
plt.xticks(rotation=45)
plt.show()
```
