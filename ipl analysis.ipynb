```python
data_mat.drop('umpire3',axis=1,inplace=True)
```

There are very less number of third umpire so its better to remove that column

```python
data_mat.isnull().sum()
```

```python
data_mat['umpire1'].value_counts()   
```

```python
data_mat.umpire1=data_mat.umpire1.fillna("HDPK Dharmasena ")
```

Filled the missing value of first umpire by HDPK Dharmasena as its number is high

```python
data_mat.info()
```

```python
data_mat['umpire2'].value_counts()
```

```python
data_mat.umpire2=data_mat.umpire2.fillna('S Ravi')
```

Filled the two missing value of second umpire by S Ravi as its number IS high

```python
data_mat['city'].value_counts()
```

Most number of matches from 2008-19 was played in Mumbai 

```python
data_mat.city=data_mat.city.fillna('-')
```

Filled the 7 NAN values of city by   '---'

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(list(data_mat['city'].value_counts()[0:10].keys()),list(data_mat['city'].value_counts()[0:10]),color='g')
plt.xlabel('CITY')
plt.ylabel('NUMBERS')
plt.title('Top 10 city with most match played')
plt.xticks(rotation=45)
plt.show()
```

# Analysis on first batting team

```python
batting_first=data_mat[data_mat['win_by_runs']!=0]
```

```python
batting_first
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'],bins=40)
plt.title('distribution by run')
plt.xlabel('Runs')
plt.ylabel('No. of matches')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color='g')
plt.xlabel('TEAM')
plt.ylabel('NUMBERS')
plt.title('Top 5 First Batting Winner')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
