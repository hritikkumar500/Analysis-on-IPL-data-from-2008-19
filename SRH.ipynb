# batting Career

```python
srh_batting=data_del[data_del['batting_team']=='Sunrisers Hyderabad']
```

```python
srh_batting
```

```python
srh_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(srh_batting['batsman_runs'].value_counts()[0:5].keys()),list(srh_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by Sunrisers')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(srh_batting['batsman_runs'].value_counts()),labels=list(srh_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
srh_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(srh_batting['dismissal_kind'].value_counts()),labels=list(srh_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
srh_bowling
```

```python
srh_bowling['bowler'].value_counts()
```

```python
srh_bowling['wide_runs'].value_counts()
```

```python
srh_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(srh_bowling['batsman_runs'].value_counts()[0:5].keys()),list(srh_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY SUNRISERS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball=srh_bowling[srh_bowling['wide_runs']!=0]
```

```python
wide_ball
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball['wide_runs'].value_counts()[0:5].keys()),list(wide_ball['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('Count Of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY SUNRISERS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```
