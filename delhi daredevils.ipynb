```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
dd_batting=data_del[data_del['batting_team']=='Delhi Daredevils']
```

```python
dd_batting
```

```python
dd_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(dd_batting['batsman_runs'].value_counts()[0:5].keys()),list(dd_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by DELHI DAREDEVILS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(dd_batting['batsman_runs'].value_counts()),labels=list(dd_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
dd_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(dd_batting['dismissal_kind'].value_counts()),labels=list(dd_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
dd_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(dd_bowling['batsman_runs'].value_counts()[0:5].keys()),list(dd_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY DELHI DAREDEVILS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```
