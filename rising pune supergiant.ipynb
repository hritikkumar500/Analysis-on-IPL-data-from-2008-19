```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
rps_batting=data_del[data_del['batting_team']=='Rising Pune Supergiant']
```

```python
rps_batting
```

```python
rps_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rps_batting['batsman_runs'].value_counts()[0:5].keys()),list(rps_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS BY RISING PUNE SUPERGIANT')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rps_batting['batsman_runs'].value_counts()),labels=list(rps_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
rps_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rps_batting['dismissal_kind'].value_counts()),labels=list(rps_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rps_bowling['batsman_runs'].value_counts()[0:5].keys()),list(rps_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY RISING PUNE SUPERGIANT BOWLERS')
plt.xticks(rotation=45)
plt.show()
```
