```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
kkr_batting=data_del[data_del['batting_team']=='Kolkata Knight Riders']
```

```python
kkr_batting
```

```python
kkr_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(kkr_batting['batsman_runs'].value_counts()[0:5].keys()),list(kkr_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by KOLKATA KNIGHT RIDERS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(kkr_batting['batsman_runs'].value_counts()),labels=list(kkr_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
kkr_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(kkr_batting['dismissal_kind'].value_counts()),labels=list(kkr_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
kkr_bowling['wide_runs'].value_counts()
```

```python
kkr_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(kkr_bowling['batsman_runs'].value_counts()[0:5].keys()),list(kkr_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY KOLKATA KNIGHT RIDERS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball_kkr=kkr_bowling[kkr_bowling['wide_runs']!=0]
```

```python
wide_ball_kkr
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_kkr['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_kkr['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('count of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY kolkata knight riders')
plt.xticks(rotation=45)
plt.show()
```
