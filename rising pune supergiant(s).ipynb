```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
rpss_batting=data_del[data_del['batting_team']=='Rising Pune Supergiants']
```

```python
rpss_batting
```

```python
rpss_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rpss_batting['batsman_runs'].value_counts()[0:5].keys()),list(rpss_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs BY RISING PUNE SUPERGIANTS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rpss_batting['batsman_runs'].value_counts()),labels=list(rpss_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
rpss_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(rpss_batting['dismissal_kind'].value_counts()),labels=list(rpss_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(rpss_bowling['batsman_runs'].value_counts()[0:5].keys()),list(rpss_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY RISING PUNE SUPERGIANTS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_rpss['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_rpss['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('Count Of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY RISING PUNE SUPERGIANTS')
plt.xticks(rotation=45)
plt.show()
```
