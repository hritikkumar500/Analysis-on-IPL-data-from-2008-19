```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
mi_batting=data_del[data_del['batting_team']=='Mumbai Indians']
```

```python
mi_batting
```

```python
mi_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(mi_batting['batsman_runs'].value_counts()[0:5].keys()),list(mi_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by MUMBAI INDIANS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(mi_batting['batsman_runs'].value_counts()),labels=list(mi_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
mi_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(mi_batting['dismissal_kind'].value_counts()),labels=list(mi_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

# Bowling career

```python
mi_bowling=data_del[data_del['bowling_team']=='Mumbai Indians']
```

```python
mi_bowling
```

```python
mi_bowling['bowler'].value_counts()
```

```python
mi_bowling['wide_runs'].value_counts()
```

```python
mi_bowling['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(mi_bowling['batsman_runs'].value_counts()[0:5].keys()),list(mi_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY MUMBAI INDIANS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball_mi=mi_bowling[mi_bowling['wide_runs']!=0]
```

```python
wide_ball_mi
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_mi['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_mi['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('Count Of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY MUMBAI UNDIANS')
plt.xticks(rotation=45)
plt.show()
```
