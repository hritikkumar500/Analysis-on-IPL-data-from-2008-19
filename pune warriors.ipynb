```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_del=pd.read_csv('deliveries.csv')
data_del
```

# batting career

```python
pw_batting=data_del[data_del['batting_team']=='Pune Warriors']
```

```python
pw_batting
```

```python
pw_batting['batsman_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(pw_batting['batsman_runs'].value_counts()[0:5].keys()),list(pw_batting['batsman_runs'].value_counts()[0:5]),color='y')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('Runs by PUNE WARRIORS')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(pw_batting['batsman_runs'].value_counts()),labels=list(pw_batting['batsman_runs'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
pw_batting['dismissal_kind'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(pw_batting['dismissal_kind'].value_counts()),labels=list(pw_batting['dismissal_kind'].value_counts().keys()),autopct='%0.1f%%',labeldistance=1.1,rotatelabels=True, pctdistance=0.8,shadow=True)
plt.show()
```

# bowling career

```python
pw_bowling=data_del[data_del['bowling_team']=='Pune Warriors']
```

```python
pw_bowling
```

```python
pw_bowling['bowler'].value_counts()
```

```python
pw_bowling['wide_runs'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(pw_bowling['batsman_runs'].value_counts()[0:5].keys()),list(pw_bowling['batsman_runs'].value_counts()[0:5]),color='b')
plt.xlabel('count of Boundries,singles and dot ball')
plt.ylabel('Numbers')
plt.title('RUNS GIVEN BY PUNE WARRIORS BOWLERS')
plt.xticks(rotation=45)
plt.show()
```

```python
wide_ball_pw=pw_bowling[pw_bowling['wide_runs']!=0]
```

```python
wide_ball_pw
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(wide_ball_pw['wide_runs'].value_counts()[0:5].keys()),list(wide_ball_pw['wide_runs'].value_counts()[0:5]),color='g')
plt.xlabel('Count Of Wide Balls')
plt.ylabel('Numbers')
plt.title('COUNT OF WIDE RUNS GIVEN BY PUNE WARRIORS')
plt.xticks(rotation=45)
plt.show()
```
